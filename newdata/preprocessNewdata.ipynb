```python
path = 'Energy_consumption_private_dwellings_24012025_092636.csv'
data = pd.read_csv(path, delimiter=';') 
```

```python
import seaborn as sns
```

```python
# Rename columns for clarity
data = data.rename(columns={
    "Periods": "Year",
    "Average consumption of natural gas (m3)": "AvgGasConsumption",
    "Average supply of electricity (kWh)": "AvgElectricitySupply"
})

# Clean and convert numerical columns
data["AvgGasConsumption"] = (
    data["AvgGasConsumption"].astype(str).str.strip().replace({".": None, "": None}).astype(float)
)
data["AvgElectricitySupply"] = (
    data["AvgElectricitySupply"].astype(str).str.strip().replace({".": None, "": None}).astype(float)
)

# Extract numeric years
data["Year"] = data["Year"].astype(str).str.extract(r"(\d+)")[0].astype(int)

# Plot gas consumption trends by region
plt.figure(figsize=(12, 6))
sns.lineplot(data=data, x="Year", y="AvgGasConsumption", hue="Regions", marker="o")
plt.xlabel("Year")
plt.ylabel("Average Gas Consumption (m3)")
plt.title("Gas Consumption Trends by Region")
plt.legend(loc="upper right", bbox_to_anchor=(1.3, 1))
plt.grid(True)
plt.show()

# Plot electricity supply trends by region
plt.figure(figsize=(12, 6))
sns.lineplot(data=data, x="Year", y="AvgElectricitySupply", hue="Regions", marker="s")
plt.xlabel("Year")
plt.ylabel("Average Electricity Supply (kWh)")
plt.title("Electricity Supply Trends by Region")
plt.legend(loc="upper right", bbox_to_anchor=(1.3, 1))
plt.grid(True)
plt.show()
```
